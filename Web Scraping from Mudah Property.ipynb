```python
from bs4 import BeautifulSoup
import pandas as pd
import requests
import json
```

```python
url = 'https://www.mudah.my/kuala-lumpur/apartment-condominium-for-rent?o=6'

source = requests.get(url).text

headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36',
}


response = requests.get(url, headers=headers)

if response.status_code == 200:
    print(response.text)  # Print the content of the response
else:
    print(f'Request failed with status code: {response.status_code}')
```

```python
soup = BeautifulSoup(response.content,'html.parser')
script = soup.find_all('script')[8].text.strip()
print(script)
```

```python
data = json.loads(script)
```

```python
data
```

```python
# Extract the relevant information
extracted_data = []
for item in data[2]['itemListElement']:
    entry = item['item']
    name = entry['name']
    description = entry['description']
    floor_size = entry['floorSize']
    price = entry['price']
    extracted_data.append({'name': name, 'description': description, 'floor_size': floor_size, 'price': price})

df = pd.DataFrame(extracted_data)
```

```python
df_filtered = df.iloc[:30]
df_filtered.to_csv('file_name.csv', encoding='utf-8')
```

```python
## Create a list of property
Property_name = ['Plaza Damas 3 Condo Sri Hartamas',
                 'Bukit Pandan Condominium',
                 'Residensi Gembira 737',
                 'Aradia Residence Lake City',
                 'Fera Residence',
                 'The Valley @ Sky Sierra',
                 '10 Stonor Residensi',
                 'M Centura',
                 'Desa Green',
                 'Bukit Pandan Kondominium 1',
                 'Bayu Sentul Condominium',
                 'The Crest condo',
                 'Condo Sri Jelatek',
                 'BUKIT PANDAN2 CONDO',
                 'INSPIRASi MONT KIARA',
                 'The Veo Melawati',
                 'PPA1M Bukit Jalil Condo',
                 'The Rosewoodz',
                 'Platinum Hill PV 3',
                 'Kenwingston Platz',
                 'Bayu Andaman',
                 'Danau Murni Condo',
                 'Nadi Bangsar Residence',
                 'MAXIM CITYLIGHTs',
                 'Ativo Suites',
                 'Armani Cheras UCSI Residence',
                 'Irama Wangsa Condominium',
                 'Residensi Enesta Kepong',
                 'Sri Putramas'
                 ]

## Create a list of area
Area_name = ['Mont Kiara',
             'Ampang',
             'Mid Valley',
             'Kepong',
             'Wangsa Maju',
             'Setiawangsa',
             'KLCC',
             'Sentul',
             'Kuala Lumpur',
             'Cheras',
             'Pandan',
             'Melawati',
             'Bukit Jalil',
             'Bangsar',
             'Gombak',
             'Setapak'
]
```

```python
def extract_propertyName(name):
    if isinstance(name, str):
        for propertyname in Property_name:
            if f"{propertyname}" in name:
                return propertyname
    return 'Undefined'

def extract_area(name):
    if isinstance(name, str):
        for area in Area_name:
            if area.lower() in name.lower():
                return area
    return 'Undefined'
```

```python
# Apply the functions to the 'name' field
df_filtered['Property Name'] = df_filtered['name'].apply(extract_propertyName)

df_filtered['Area'] = df_filtered['description'].apply(extract_area)
```

```python
columns = ['Property Name', 'Area', 'floor_size', 'price']

New_data = df_filtered[columns]
```

```python
New_data
```

```python
New_data.to_csv('output.csv', encoding='utf-8')
```

```python
# Convert the 'price' column to string
New_data['price'] = New_data['price'].astype(str)

# Remove "RM " from the price column and convert it to numeric, handling errors
New_data['price'] = New_data['price'].str.replace('RM ', '').str.replace(',', '').str.strip()
New_data['price'] = pd.to_numeric(New_data['price'], errors='coerce')

# Convert floor_size to numeric, handling errors
New_data['floor_size'] = pd.to_numeric(New_data['floor_size'], errors='coerce')

# Drop rows with NaN values in 'floor_size' or 'price'
New_data = New_data.dropna(subset=['floor_size', 'price'])

# Print the cleaned data to verify
print(New_data.head())
```

```python
New_data
```

```python
# Group by Property Name and calculate average floor size and average price
average_data = New_data.groupby(['Property Name','Area']).agg(
    average_floor_size=('floor_size', 'mean'),
    average_rental=('price', 'mean')
).reset_index()
```
